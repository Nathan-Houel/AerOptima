## **Initialisation**

```python
import pandas as pd
```

```python
# Data Loading
flights = pd.read_csv("Dataset/flights.csv")
airports = pd.read_csv("Dataset/airports.csv")
airlines = pd.read_csv("Dataset/airlines.csv")
```

```python
# Display the columns
print(flights.columns)
```

```python
# Display the first 5 lines
flights.head(5)
```

```python
# Display the columns
print(airlines.columns)
```

```python
# Display the first 5 lines
airlines.head(5)
```

```python
# Joining flights and airlines
flights_merged = flights.merge(airlines, left_on='AIRLINE', right_on='IATA_CODE')
flights_merged.head(5)
```

```python
# Renaming AIRLINE_y to AIRLINE
flights_merged.rename(columns={'AIRLINE_y': 'AIRLINE'}, inplace=True)

# Deleting AIRLINE_x
flights_merged.drop(columns=['AIRLINE_x'], inplace=True)
flights_merged.head(5)
```

## **Temporal puzzle** ‚è≥

```python
flights_merged[['YEAR','MONTH','DAY','SCHEDULED_DEPARTURE']].head(10)
```

```python
# Converting columns SCHEDULED_DEPARTURE to string
flights_merged['SCHEDULED_DEPARTURE'] = flights_merged['SCHEDULED_DEPARTURE'].apply(str)
print(flights_merged['SCHEDULED_DEPARTURE'])
```

```python
# Adding '00' to the rest of the number
flights_merged['SCHEDULED_DEPARTURE'] = flights_merged['SCHEDULED_DEPARTURE'].str.zfill(4)
flights_merged.head(5)
```

```python
# Create a new columns with full date in string
flights_merged['DATE_STR'] = flights_merged['YEAR'].astype(str) + '-' + flights_merged['MONTH'].astype(str) + '-' + flights_merged['DAY'].astype(str)
flights_merged.head(5)
```

```python
# Create a new column to merge date with time
flights_merged['STR_TIMESTAMP'] = flights_merged['DATE_STR'] + ' ' + flights_merged['SCHEDULED_DEPARTURE']
flights_merged.head(5)
```

```python
# Creating timestamp column
flights_merged['TIMESTAMP'] = pd.to_datetime(flights_merged['STR_TIMESTAMP'], format = '%Y-%m-%d' + ' ' + '%H%M')
flights_merged.info()
```

```python
# Deleting columns DATE_STR and STR_TIMESTAMP
flights_merged.drop(columns=['DATE_STR', 'STR_TIMESTAMP'], inplace=True)
```

```python
# Displaying all columns's name + creating new order of columns
cols = flights_merged.columns.tolist()
new_cols = cols[:4] + cols[-1:] + cols[-2:-1] + cols[4:-2]
print(new_cols)
```

```python
# Creating the clean version
flights = flights_merged[new_cols]
flights.head(5)
```
